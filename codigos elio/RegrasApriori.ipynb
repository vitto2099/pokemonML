# Base mercado toy

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transacoes = pd.read_csv("mercado.csv",header=None)
transacoes
```

```python
transacoes = transacoes.apply(lambda x: [item for item in x if pd.notnull(item)], axis=1)
transacoes
```

```python
transactionEncoder = TransactionEncoder()
transacoes_ary = transactionEncoder.fit(transacoes).transform(transacoes)
df_transacoes = pd.DataFrame(transacoes_ary , columns=transactionEncoder.columns_)
df_transacoes
```

```python
frequent_itemsets = apriori(df_transacoes, min_support=0.30, use_colnames=True)
```

```python
regras = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.80)
```

```python
regras = regras[(regras['confidence'] >= 0.8) & (regras['lift'] >= 2)]
regras
```

```python
regras.to_csv("regras.csv", sep=",", index=False)
```

# Base mercado

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

transacoes = pd.read_csv("mercado2.csv")
transacoes

```

```python
transacoes = transacoes.apply(lambda x: [item for item in x if pd.notnull(item)], axis=1)

transactionEncoder = TransactionEncoder()
transacoes_ary = transactionEncoder.fit(transacoes).transform(transacoes)
df_transacoes = pd.DataFrame(transacoes_ary , columns=transactionEncoder.columns_)


frequent_itemsets = apriori(df_transacoes, min_support=0.01, use_colnames=True)

regras = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)

regras = regras[(regras['confidence'] >= 0.30)]

regras = regras[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
regras

regras.to_csv("regrasMercadoCompleto.csv", sep=",", index=False)
```

### Subamostragem com Tomek Links

```python
from imblearn.under_sampling import TomekLinks
```

```python
tl = TomekLinks(sampling_strategy='majority') #all
X_under, y_under = tl.fit_resample(X_credit, y_credit)
```

```python
X_under.shape, y_under.shape
```

```python
np.unique(y_credit, return_counts=True)
```

```python
np.unique(y_under, return_counts=True)
```

### Sobreamostragem com SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X_credit.shape
```

```python
smote = SMOTE(sampling_strategy='minority')
X_over, y_over = smote.fit_resample(X_credit, y_credit)
```

```python
X_over.shape
```

```python
np.unique(y_credit, return_counts=True)
```

```python
np.unique(y_over, return_counts=True)
```
